## рҹ“• Course Contents

The course is divided into two parts.

---

### Part I вҖ” Interactive Data Visualization with Plotly рҹ“Ҡ
*(Lectures 1вҖ“8)*

We move from exploratory charts to professional, explanatory visualizations вҖ” applying design principles from *Storytelling with Data* (Knaflic) throughout.

| Lecture | Topic | Chart Types | Dataset |
|---|---|---|---|
| L1 | Course intro + Plotly basics | вҖ” | Stocks (built-in) |
| L2 | Bar charts & design principles | Bar, horizontal bar, stacked | World Happiness Report |
| L3 | Line charts & slopegraphs | Line, slopegraph | COвӮӮ Emissions |
| L4 | Scatter & bubble charts | Scatter, bubble | Gapminder |
| L5 | Distribution charts | Histogram, box, violin | Airbnb London |
| L6 | Hierarchical charts | Treemap, sunburst | Global Energy Mix |
| L7 | Heatmaps & waterfall charts | Heatmap, waterfall | Netflix Catalogue |
| L8 | Choropleth maps + full story | Choropleth, multi-chart story | World Development |

**Part I learning outcomes:**
- Apply SWD chart-selection and design principles to real datasets
- Build professionally styled, insight-driven Plotly charts
- Use grey-and-highlight colour technique, zero baseline, insight titles
- Structure a multi-chart narrative 

---

### Part II вҖ” Building Interactive Dashboards with Streamlit рҹҡҖ
*(Lectures 9вҖ“14)*

We take the charts from Part I and turn them into multi-page, filterable, deployable dashboards вҖ” applying design principles from *The Big Book of Dashboards* (Wexler, Shaffer & Cotgreave).

| Lecture | Topic | Key Skills |
|---|---|---|
| L9  | Streamlit foundations | `st.metric`, `st.columns`, `st.sidebar`, `st.plotly_chart`, deploy |
| L10 | Filters & interactivity | `st.selectbox`, `st.multiselect`, `st.slider`, `st.date_input`, chained filters |
| L11 | Multi-page apps | `st.navigation`, `pages/`, `utils.py`, `st.session_state`, `st.tabs` |
| L12 | Design & polish | Custom CSS, `@st.cache_data`, BBD colour rules, 5-second test |
| L13 | Capstone build session | Guided in-class dashboard build  |
| L14 | Open Q&A | final questions on course material & final project  |

**Part II learning outcomes:**
- Build multi-page Streamlit dashboards from scratch
- Apply BBD filter design, colour taxonomy (sequential/diverging/categorical/highlight)
- Handle CVD accessibility (blue-orange instead of red-green)
- Deploy a working dashboard to Streamlit Community Cloud
- Present a data product with a clear Big Idea and audience
